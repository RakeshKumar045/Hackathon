```python
xgb_preds = pd.read_csv('/kaggle/input/predictions-av-bank-data/xgb_annual.csv')
cat_preds = pd.read_csv('/kaggle/input/predictions-av-bank-data/cat_annual.csv')
lgb_preds = pd.read_csv('/kaggle/input/lgb-sol/lgb_annual.csv')

final_preds = pd.DataFrame(columns = xgb_preds.columns)
final_preds
```

```python
final_preds['Loan_ID'] = xgb_preds['Loan_ID']
x = pd.concat([xgb_preds['Interest_Rate'], cat_preds['Interest_Rate'], lgb_preds['Interest_Rate']], axis = 1)
```

```python
final_preds['Interest_Rate'] = np.median(x, axis = 1)
final_preds
```
